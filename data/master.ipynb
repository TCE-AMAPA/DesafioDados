```python
import pandas as pd
from sqlalchemy import create_engine

tabela=pd.read_csv("rgf_anx_01.csv")

# Fiz o import da tabela e a leitura inicial do arquivo csv

colunas_int = ["periodo", "exercicio", "cod_ibge", "populacao"]

for coluna in colunas_int:
    tabela[coluna] = pd.to_numeric(tabela[coluna], errors='coerce').astype(int)

tabela["valor"] = tabela["valor"].astype(float)

tabela["rotulo"] = tabela["rotulo"].fillna("Nao informado").astype(str)

tabela["_extracted_at"] = pd.to_datetime(tabela["_extracted_at"])
# Converti as colunas que tinham numeros inteiros para type int, a coluna valor para float, a coluna rotulo para str e a coluna _extracted_at para data.

CONN_STR = "postgresql://usuario_teste:senha_forte@localhost:5432/rgf_dados"

engine = create_engine(CONN_STR)
# Criei a engine do banco.

try:
    tabela.to_sql(
        name='tabela_tratada', 
        con=engine, 
        if_exists='replace',          
        index=False,                  
        method='multi',               
        chunksize=1000                
    )
    print("Dados inseridos com sucesso")
except Exception as e:
    print(f"Erro ao inserir dados: {e}")
# Mandei a tabela para o banco.

display(tabela)
```

```python
try:
    query = "SELECT * FROM tabela_tratada LIMIT 10" # O preview das 10 linhas.
    tabela_do_banco = pd.read_sql(query, con=engine) # Usando a engine pra criar a váriavel da tabela no banco de dados.
    
    print("Dados recuperados do banco com sucesso:")
    display(tabela_do_banco.style.format(
        {
            'valor': 'R$ {:,.2f}'.format,
            'populacao': '{:,.0f}'.format 
        },
        decimal=',',
        thousands='.'
    )) # Mostrar a tabela diretamente do banco de dados, com filtro para mostrar o float no padrão br e a população em formato de milhar.
    
    print("\nTipos de colunas no DataFrame vindo do banco:")
    print(tabela_do_banco.dtypes) # usando dtypes para printar os tipos das colunas.
    
except Exception as e:
    print(f"Erro ao ler dados do banco: {e}")
```
